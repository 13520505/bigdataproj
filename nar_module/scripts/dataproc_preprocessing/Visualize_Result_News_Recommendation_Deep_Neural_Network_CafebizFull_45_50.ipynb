# Biểu đồ hiển thị giá trị Hit Rate theo giờ của từng thuật toán

```python
pd_benchmark = pd.read_csv("/data/tungtv/Code/dataset/dataset_cafebiz_full_45/tmp/chameleon/jobs/tungtv_cafebiz_2019_07_19_203900/eval_stats_benchmarks.csv")
pd_hitrate = pd.DataFrame(columns=['index','hitrate_at_n_cb'\
                                   ,'hitrate_at_n_chameleon','hitrate_at_n_coocurrent'\
                                   ,'hitrate_at_n_item_knn','hitrate_at_n_pop_recent',\
                                   'hitrate_at_n_sr','hitrate_at_n_v-sknn'])

pd_hitrate['index'] = pd_benchmark['index'] 
pd_hitrate['hitrate_at_n_cb'] = pd_benchmark['hitrate_at_n_cb']
pd_hitrate['hitrate_at_n_chameleon'] = pd_benchmark['hitrate_at_n_chameleon']
pd_hitrate['hitrate_at_n_coocurrent'] = pd_benchmark['hitrate_at_n_coocurrent']
pd_hitrate['hitrate_at_n_item_knn'] = pd_benchmark['hitrate_at_n_item_knn']
pd_hitrate['hitrate_at_n_pop_recent'] = pd_benchmark['hitrate_at_n_pop_recent']
pd_hitrate['hitrate_at_n_sr'] = pd_benchmark['hitrate_at_n_sr']
pd_hitrate['hitrate_at_n_v-sknn'] = pd_benchmark['hitrate_at_n_v-sknn']


col_hitrate = ['hitrate_at_n_chameleon','hitrate_at_n_cb',\
        'hitrate_at_n_coocurrent'\
        ,'hitrate_at_n_item_knn','hitrate_at_n_pop_recent'\
        ,'hitrate_at_n_sr','hitrate_at_n_v-sknn']
lables = ['1','2','3','4','5','6','7']

import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.xlabel("Hour")
plt.ylabel("Hit Rate")
plt.title("Average HR@5 by hour Cafebiz in April and May")

lines = pd_hitrate.plot.line(ax = plt.gca(),x='index', y=col_hitrate, grid=True)
lines

plt.savefig("hitrate_cafebiz_45_50.png")

```

# Biểu đồ hiển thị giá trị MMR theo giờ của từng thuật toán

```python
pd_mmr = pd.DataFrame(columns=['index',\
       'mrr_at_n_cb', 'mrr_at_n_chameleon', 'mrr_at_n_coocurrent',\
       'mrr_at_n_item_knn', 'mrr_at_n_pop_recent', 'mrr_at_n_sr',\
       'mrr_at_n_v-sknn'])

pd_mmr['index'] = pd_benchmark['index'] 
pd_mmr['mrr_at_n_cb'] = pd_benchmark['mrr_at_n_cb']
pd_mmr['mrr_at_n_chameleon'] = pd_benchmark['mrr_at_n_chameleon']
pd_mmr['mrr_at_n_coocurrent'] = pd_benchmark['mrr_at_n_coocurrent']
pd_mmr['mrr_at_n_item_knn'] = pd_benchmark['mrr_at_n_item_knn']
pd_mmr['mrr_at_n_pop_recent'] = pd_benchmark['mrr_at_n_pop_recent']
pd_mmr['mrr_at_n_sr'] = pd_benchmark['mrr_at_n_sr']
pd_mmr['mrr_at_n_v-sknn'] = pd_benchmark['mrr_at_n_v-sknn']

print(pd_mmr['mrr_at_n_cb'].mean())
print(pd_mmr['mrr_at_n_chameleon'].mean())
print(pd_mmr['mrr_at_n_coocurrent'].mean())
print(pd_mmr['mrr_at_n_item_knn'].mean())
print(pd_mmr['mrr_at_n_pop_recent'].mean())
print(pd_mmr['mrr_at_n_sr'].mean())
print(pd_mmr['mrr_at_n_v-sknn'].mean())

col_mmr=[ 'mrr_at_n_chameleon', 'mrr_at_n_cb', 'mrr_at_n_coocurrent',\
       'mrr_at_n_item_knn', 'mrr_at_n_pop_recent', 'mrr_at_n_sr',\
       'mrr_at_n_v-sknn']

import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.xlabel("Hour")
plt.ylabel("MMR")
plt.title("Average MRR@5 by hour Cafebiz in  April and May")

lines = pd_mmr.plot.line(x='index', y=col_mmr, ax = plt.gca(), grid=True)
lines

# plt.savefig("mrr_cafebiz_45.png")
```
